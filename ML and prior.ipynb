```python
# Sklearn Preprocessing
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler
# Sklearn Modeling
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.ensemble import VotingClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
```

```python
X = churn.drop('churn', axis = 1)
y = churn['churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 30)
```

```python
num_cols = X_train.select_dtypes('number').columns
cat_cols = X_train.select_dtypes('object').columns
```

```python
print(num_cols,cat_cols)
```

```python
num_transformer = Pipeline(steps = [('scaler', StandardScaler())])
cat_transformer = Pipeline(steps = [('encoder', OneHotEncoder(handle_unknown = 'ignore', sparse = False))])
preprocessing = ColumnTransformer(transformers = [('num', num_transformer, num_cols),
                                                  ('cat', cat_transformer, cat_cols)])
X_train_trans = preprocessing.fit_transform(X_train)
X_test_trans = preprocessing.transform(X_test)
```

For the sake of keeping things simple, we’ll take the extra step to put our processed data back into a pandas dataframe.

```python
slice_pipe = preprocessing.named_transformers_['cat']
cat_features = slice_pipe.named_steps['encoder'].get_feature_names_out(cat_cols)
```

```python
print(cat_features)
```

```python
X_train_df = pd.DataFrame(X_train_trans, columns = [*num_cols, *cat_features])
X_test_df = pd.DataFrame(X_test_trans, columns = [*num_cols, *cat_features])
```

```python
# Finding the best hyperparameters
estimator = LogisticRegression(class_weight = 'balanced')
params = {'penalty': ['l1', 'l2', 'elasticnet', 'None'],
          'C': [0.5, 1, 2],
          'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
          'max_iter': [100, 150, 200],
          'multi_class': ['auto', 'ovr', 'multinomial']}
# Fit the gridsearch 
grid = GridSearchCV(estimator = estimator,
                    param_grid = params,
                    cv = 3, 
                    n_jobs = -1)
jit(target_backend='cuda',nopython=True)
grid.fit(X_train_df, y_train)
best_logreg_params = grid.best_params_
best_logreg_model = LogisticRegression(**best_logreg_params, class_weight = 'balanced')
best_logreg_model.fit(X_train_df, y_train)
```

```python
# from numba import jit, cuda
jit(target_backend='cuda',nopython=True)
def evaluate_model(model, X_test, y_test):
    '''Evaluates a sci-kit learn classification model and displays a classification report, 
    a confusion matrix, and a ROC-AUC plot.'''
    
    # Predictions
    y_hat_test = model.predict(X_test)
    y_hat_test_proba = model.predict_proba(X_test)
    cm = confusion_matrix(y_test, y_hat_test, labels=model.classes_)
    
    # Classification Report
    print('                 Classification Report')
    print('-------------------------------------------------------')
    fpr, tpr, thresholds = metrics.roc_curve(y_test, y_hat_test)
    print(classification_report(y_test, y_hat_test))
    roc_auc = metrics.auc(fpr, tpr)
    # Confusion Matrix Plot
    fig, axes = plt.subplots(figsize = (12, 6), ncols = 2)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    #plt.plot([0, 1], [0, 1], 'k--', label = 'Random')
    #plt.plot(fpr, tpr, color = 'blue', label = 'ROC')
    #plt.xlabel('False Positive Rate', fontsize=15)
    #plt.ylabel('True Positive Rate', fontsize=15)
    #plt.title('ROC Curve', fontsize=20)
    skplt.metrics.plot_roc(y_test, y_hat_test_proba,ax=axes[1])
    plt.legend(fontsize=10, loc='best')
    disp.plot(ax=axes[0])
    # ROC-AUC Curve
    # roc_auc_plot = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc)
    # axes[1].legend()
    # axes[1].plot([0,1], [0,1], ls = ':')
    # axes[1].grid()
    # roc_auc_plot.plot()
    # axes[1].set_title('ROC-AUC Plot')
    # fig.tight_layout()
    plt.show();
```

```python
# Evaluate model
# jit(target_backend='cuda',nopython=True)
evaluate_model(best_logreg_model, X_test_df, y_test)
```

We can use the SHAP package to not only see which features the model deemed to be important in predicting customer churn, but we can also see whether each feature affects churn positively or negatively. -->
BUT we’re not able to see whether the factors have a positive or negative effect. For that, we can use another package called SHAP to get a more clear picture of what’s happening here.

```python
# Setting up SHAP
X_shap = shap.sample(X_train_df, 100)
#explainer = shap.TreeExplainer(best_xgb_model,X_train_df)
#shap_values_Tree_tr = shap.TreeExplainer(best_xgb_model).shap_values(X_shap)
explainer= shap.explainers.Tree(best_xgb_model)
#shap.summary_plot(shap_values_Tree_tr, X_train)
shap_values = explainer.shap_values(X_shap)
# Another shap plot that shows how each feature affects churn
shap.summary_plot(shap_values, X_shap)
```

Hence Total day minutes is having much more negative impact according to data points compared to positive impact.
Total day minutes, customer service calls, not having international plan , total international calls and total evenings minutes. Whereas total day minutes, total international minutes and total evening minutes have a noticable positive impact also.
